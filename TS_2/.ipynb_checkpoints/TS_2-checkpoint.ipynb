## Introducción

La combinación de distintos elementos eléctricos en un circuito permite la conformación de lo que se conoce como **filtros elécticos o electrónicos**. Un filtro es un dispositivo capaz de discriminar una frecuencia, o una gama de ellas, provenientes de una señal que lo atraviese, la cual, puede ser modificada tanto en amplitud como en fase por este dispositivo.

En el siguiente trabajo se procederá al análisis de dos circuitos eléctricos donde se buscará conocer sus respectivas funciones de transferencia. A partir de estos resultados, se representarán las respuestas en módulo y fase respecto a la frecuencia en cada circuito, con el fin de poder caracterizar la clase de filtro electrónico pasivo de la cual se trata como, a su vez, notar las características principales que genera la variación de sus parámetros.

### Caso A

El circuito en cuestión es:

<div style="text-align: center;">
    <img src="./pasabanda.jpg" width="500">
</div>
Para el cual se realizó el siguiente análisis:

```python
from IPython.display import IFrame
IFrame("./ts2_1.pdf", width=1000, height=1000)
```

Y los gráficos obtenidos fueron:

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.logspace(-5,5,10000) # Vector de frecuencia en escala logarítmica
wo = 1
Q = 1/np.sqrt(2) # Factor de selectividad

def pasa_banda_rlc(w,wo,Q):
    

    s = 1j * w
    
    H = (wo/Q) * s/((wo/Q)*s + wo**2 + s**2) # Función de transferencia del pasabanda 
       
    escallog = 20*np.log10(np.abs(H)) # Módulo de la función de transeferencia en escala logarítmica (resultado en decibeles)
    
    fase = np.angle(H,False) # Fase de la función de transferencia
        
    plt.figure(1)
    plt.title("Respuesta en módulo - Pasa Banda")
    plt.xlabel("w [Rad/s]")
    plt.ylabel("|H|[dB]")
    plt.semilogx(w,escallog)
    plt.grid(True)
    
    plt.figure(2)
    plt.title("Respuesta en Fase - Pasa Banda")
    plt.xlabel("w [Rad/s]")
    plt.ylabel("Fase [Rad]")
    plt.semilogx(w,fase)
    plt.grid(True)
    
    return

pasa_banda_rlc(w,wo,Q)  
```

### Resultados - Circuito A

Obtenida analíticamente la función de transferencia de este cicuito y su respectivo comportamiento en fase y en módulo, se aprecia que:
- En módulo, a bajas frecuencias, la ganancia tiende a comportarse como una asintota lineal de pendiente $+20\frac{dB}{década}$
- En módulo, a altas frecuencias, la ganancia tiende a comportarse como una asintota lineal de pendiente $-20\frac{dB}{década}$
- En la respuesta de módulo, la ganancia es máxima al corresponderse con la frecuencia $\omega 0$
- El ancho de banda, en este caso, resulta de $\sqrt{2}$ al analizar el gráfico la distancia entre las dos frecuencias de corte. Resultado que se corresponde con la fórmula estipulada para calcular el ancho de banda para un pasa bandas (en este caso con $\omega 0 = 1$ y $Q = \frac{1}{\sqrt{2}}$)
- La respuesta en fase se despliega entre $\frac{\pi}{2} \, rad$ y $\frac{-\pi}{2} \, rad$, con un valor de 0 en $\omega 0$. Hecho también previsto analiticamente para un circuito de este tipo.
  
Estos resultados dan indicio de que el circuito se trata de un **filtro pasa banda**, donde las frecuencias mayores o menores al ancho de banda preestablecido son atenuadas.

## Caso B

El circuito en cuestión es:

<div style="text-align: center;">
    <img src="./pasaaltos.jpg" width="500">
</div>
Para el cual se realizó el siguiente análisis:

```python
from IPython.display import IFrame
IFrame("./ts2_2.pdf", width=1000, height=1000)
```

Y los gráficos obtenidos fueron:

```python

def pasa_altos_2do_orden (w,wo,Q):
    

    s = 1j * w
    
    H = s**2 /(s**2 + wo**2 + s*(wo/Q)) # Función de transferencia del pasaaltos de segundo orden
    
    escallog = 20*np.log10(np.abs(H))
    
    fase = np.angle(H,False)
    
    plt.figure(3)
    plt.title("Respuesta en módulo - Pasa Altos 2do Orden")
    plt.xlabel("w [Rad/s]")
    plt.ylabel("|H| [dB]")
    plt.semilogx(w,escallog)
    plt.grid(True)
    
    plt.figure(4)
    plt.title("Respuesta en fase - Pasa Altos 2do Orden")
    plt.xlabel("w [Rad/s]")
    plt.ylabel("Fase [Rad]")
    plt.semilogx(w,fase)
    plt.grid(True)
    
    return
pasa_altos_2do_orden(w,wo,Q)
   
```

### Resultados - Circuito B

Obtenida analíticamente la función de transferencia de este cicuito y su respectivo comportamiento en fase y en módulo, se aprecia que:
- En módulo, a bajas frecuencias, la ganancia tiende a comportarse como una asintota lineal de pendiente $+40\frac{dB}{década}$
- En módulo, a altas frecuencias, la ganancia tiende a 0.
- En este caso, la frecuencia de corte (aquella para la cual se presenta una caida de $-3 dB$ respecto de la ganancia pasante del filtro) se situa en $\omega 0$ (que vale 1), lo cual se corresponde con la fórmula despejada para calcular la frecuencia de corte de una pasa altos de segundo orden en función de $\omega 0$ y $Q$ (que en este caso vale $\frac{1}{\sqrt{2}}$) despejada analiticamente.
- La respuesta en fase se presenta entre $\pi \,rad$ y $0  \,rad$, con un valor de $\frac{\pi}{2} \,rad$ en $\omega 0$, como se planteo analiticamente que sería el comportamiento para este tipo de circuito.
  
Estos resultados dan indicio de que el circuito se trata de un **filtro pasa altos de segundo orden**, donde las frecuencias  menores a la frecuencia de corte son atenuadas.

## Cambios en los parámetros

Al modificarse $\omega 0$ y $Q$, se obtienen los siguiente resultados gráficos: 

### Con $\omega 0$ = 1 y $Q$ = 20

```python
pasa_banda_rlc(w,wo=1,Q=20) 
```

```python
pasa_altos_2do_orden(w,wo=1,Q=20)
```

### Con $\omega 0$ = 100 y $Q$ = $\frac{1}{\sqrt{2}}$

```python
pasa_banda_rlc(w,wo=100,Q=1/np.sqrt(2)) 
```

```python
pasa_altos_2do_orden(w,wo=100,Q=1/np.sqrt(2))
```

### Resultados - Modificaciones Parámetros

- En el caso del pasa bandas, la modificación de $Q$ hacia valores mayores con respecto al que se tomo como referencia, produce una respuesta en módulo con un crecimiento más "puntiagudo" al aproximarse a la ganancia máxima. Con lo cual, se obtiene un menor ancho de banda.
- En el caso del pasa altos, la modificación de $Q$ hacia valores mayores con respecto al que se tomo como referencia, produce un pico de ganancia (sobrepasando aquella máxima vista para el caso tomado de referencia) al encontrarse en la frecuencia $\omega 0$. Con lo cual, se ve alterado también el valor de la frecuencia de corte.
- En ambos casos, el aumento de $Q$ lleva a una respuesta en fase con una transición más abrupta entre los valores límites.
- En ambos casos, la variación de $\omega 0$ produce un desplazamiento a lo largo del eje de frecuencias tanto de las respues en fase como en módulo.
- Para el pasa bandas, el aumento en el valor de $\omega 0$ conlleva un crecimiento del ancho de banda.
- Dado que se preservó el valor de $Q$ establecido originalmente, el cambio en $\omega 0$ para el pasa altos implica la misma modificación para la frecuencia de corte, conservando su igualdad respecto de $\omega 0$.

# Conclusiones

Del análisis realizado sobre los dos circuitos eléctricos y la exploración de las gráficas de las respuestas en fase y módulo de la función de transferencia en frecuencia de ambos circuitos, fue posible determinar que el circuito A se corresponde con un pasa bandas, mientras que el circuito B se trata de un filtto pasa altos de segundo orden.
Se caracterizó el comportamiento típico de cada filtro: 
- Pasa Bandas: Permite el paso de las frecuencias dentro del ancho de banda y atenua aquellas que lo excedan, lo cual se visualiza en la respuesta  de módulo; presenta un comportamiento asintótico lineal en la respuesta de módulo para las frecuencias $\omega \to 0$ y $\omega \to +\infty$ con una pendiente de $+20\frac{dB}{década}$ y $-20\frac{dB}{década}$, respectivamente; la respuesta en fase se encuentra entre  $\frac{\pi}{2} \, rad$ y $\frac{-\pi}{2} \, rad$.
- Pasa Altos Segundo Orden: Permite el paso de frecuencias superiores a la frecuencia de corte y atenua aquellas que están por debajo, representado en la respuesta en módulo de la función de transferencia; para frecuencias en las cuales $\omega \to 0$, la respuesta en módulo tiene un comportamiento asintótico lineal con una pendiente de $+40\frac{dB}{década}$; la respuesta en fase se encuentra entre  $\pi \, rad$ y $0 \, rad$.

Al modificar los parámetros $\omega 0$ y $Q$ (factor de selectividad) pudo corroborarse, como se planteo analíticamente desde la función de transeferencia de cada circuito, que: 
1) El ancho de banda del pasa bandas presenta una dependencia directamente proporcional a $\omega 0$ e inversamente con $Q$. Con lo cual, el incremento de $Q$ permitirá generar un filtro de menor ancho de banda y, por ende, más selectivo o específico en el rango de frecuencias que permitirá pasar.
2) El corrimiento de $\omega 0$ conlleva, en ambos filtros, un desplazamiento de sus respuestas (en módulo y fase) a lo largo del eje de frecuencias, corriendo con ello el ancho de banda.
3) Para el pasa altos, de coincidir $\omega $ con $\omega 0$, el valor de la ganancia de la respuesta en módulo será equivalente al valor de $Q$. Por lo tanto, valores elevados de $Q$ generan "picos" de ganancia. A su vez, la variación tanto de $\omega 0$ como de $Q$ generan modificaciones en la frecuencia de corte del pasa altos y, con ello, de las frecuencias con las cuales el filtro será permisivo.
