En el siguiente trabajo se presenta, en primera insatncia, una función generadora señales de tipo senoidal, tomando como fórmula de base:
\begin{equation}
A \,\sin(2 \,\pi \,\omega \,t + \phi) + b
\end{equation}
Con dicha función se procederá a analizar las gráfiucas obtenidas y modificar los parámetros en juego para notar las repercusiones existentes sobres las señales.

Luego se utilizará un segunda función, en este caso generadora de señales periódicas triangulares extraída de [aquí](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.sawtooth.html#scipy.signal.sawtooth).


# Código

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Mar  9 20:28:36 2025

@author: Tobi
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def funcion_senoidal (Vmax, dc, ff, ph, nn, fs):
    
    # Vmax = Amplitud máxima de la senoidal [V]
    # ff = Frecuencia [Hz]
    # fs = Frecuencia de muestreo [Hz]
    # tt = Tiempo [s]
    # ph = Fase [Rads]
    # dc = Valor medio [V]
    # nn = Cantidad de muestras
    
    Ts = 1/fs # Calculo período de muestreo
    tt = np.linspace(0,(nn-1)*Ts,nn)
    xx = Vmax * np.sin(2 * np.pi * ff * tt + ph) + dc
    
    plt.figure(1)
    plt.plot(tt,xx)
    plt.title('Señal: Senoidal')
    plt.xlabel('tiempo [s]')
    plt.ylabel('Amplitud [V]')
    plt.legend(['ff = {:d} Hz'.format(ff)],loc='upper right')
    plt.grid(True)
    
    return

def funcion_triangular(fs, n, ff, vmax, ph, dc):
    
    # vmax = Amplitud máxima de la senoidal [V]
    # ff = Frecuencia [Hz]
    # fs = Frecuencia de muestreo [Hz]
    # t = Tiempo [s]
    # ph = Fase [Rads]
    # dc = Valor medio [V]
    # n = Cantidad de muestras
    
    ts = 1/fs # Calculo período de muestreo
    t = np.linspace(0, (n-1)*ts,n)
    x = vmax * signal.sawtooth(2 * np.pi * t * ff + ph, width=0.5) + dc # width define el ancho de la rampa ascendente como proporción total del ciclo. Al ser 0.5, ambas rampas (ascendente y descendente) tienen igual ancho, definiendo un triángulo
    
    plt.figure(2)
    plt.plot(t,x, color='red')
    plt.title('Señal: Triangular')
    plt.xlabel('tiempo [s]')
    plt.ylabel('Amplitud [V]')
    plt.legend(['ff = {:d} Hz'.format(ff)],loc='upper right')
    plt.grid(True)
    
    return

N = 1000 # Cantidad de muestras a analizar
Fs = 1000
funcion_senoidal(1, 0, 1, 0, N, Fs)
#funcion_triangular(Fs, N, 5,1,0,0)

```

El gráfico expresa una señal senoidal de frecuencia 1 Hz, con amplitud de 1 V. Se denota que se cumple un solo ciclo (período) como era esperado al estar tarbajando con esta señal, una frecuencia de muestreo = 1000 Hz y una cantidad de muestras = 1000 (1000 muestras por ciclo es lo que se esperaba). En este caso, la resolución espectral ($\Delta f = \frac{Fs}{N}$) tiene valor de 1, coincidiendo con la frecuencia de la señal.

# Resultado con frecuencia = 500 Hz.

```python
funcion_senoidal(1, 0, 500, 0, N, Fs)
```

Al variar la frecuencia de la señal de forma tal que sea igual a la mitad de la frecuencia de muestreo, donde comienza a verse un error acumulativo en la señal que implica un crecimiento aparente en su amplitud a lo largo del tiempo.

# Resultado con frecuencia = 999 Hz.

```python
funcion_senoidal(1, 0, 999, 0, N, Fs)
```

Cuando la frecuencia de la señal se encuentra en $F0 = Fs - 1$, el gráfico representado se asemeja al obtenido en el primer caso, con las excepción de estar en contrafase. Este hecho plantea la existencia de una simetría de la señal respecto del valor $\frac{Fs}{2}$ dado que, en valor de frecuencia, 1 Hz y 999 Hz equidistan de 500 Hz.

# Resultado con frecuencia = 1001 Hz.

```python
funcion_senoidal(1, 0, 1001, 0, N, Fs)
```

En este caso, el gráfico obtenido es exactamente igual al del primer caso, con la salvedad que la frecuencia de la señal ahora es de 1001 Hz. Esto trae a colación la presencia una periodicidad de la representación con período igual a $Fs$.

# Resultado con frecuencia = 2001 Hz.

```python
funcion_senoidal(1, 0, 2001, 0, N, Fs)
```

En este último ejemplo se reafirma lo supuesto en el anterior: se obtiene el mismo gráfico que en el caso original a una frecuencia de señal que equivale a dos veces la frecuencia de muestreo.

# Función generadora de señales traingulares:

```python
funcion_triangular(Fs, N, 5,1,0,0)
```

Aquí se representa una señal triangular con frecuencia 5 Hz, frecuencia de muestreo = 1000 y un total de 1000 muestras. La amplitud de la señal es de 1 V.

# Análisis

Notando los resultados gráficos obtenidos de la señal senoidal pueden remarcarse varios puntos:
1) Es posible encontrarse con amibgüedades como la imposibilidad de diferenciar dos ondas cuando la frecuencia de ellas es equivalente a un múltiplo de la frecuencia de muestreo. Dando esto por hehco que la señal tendrá una periodicidad de Fs (frecuencia de muestreo).
2) Cuando la frecuencia de la señal es igual a la mitad de la frecuencia de muestreo, la señal se torna errática de manera acumulativa. Esta frecuencia particular es conocida como frecuencia de Nyquist.
3) Otro tipo de ambigüedad se presenta a partir de la frecuencia de Nyquist dado que la señal presentará una simetría especular en ella.
